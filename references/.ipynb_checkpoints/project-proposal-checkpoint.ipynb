## Introduction

Diabetic retinopathy is a serious illness, which is expected to affect > 200 million people by the year 2025 [1]. It is an eye desease resulting in blindness for over 10000 people with diabetes per year [2]. In order to help these patients Boehringer Ingelheim is investing in Research and Development of biopharmaceuticals and is screening for new active ingredients, which has the potential to slow or even stop the progression of this desease [3]. A unique characteristic of these medications is the intravitreal application, which means that the drug product is injected directly into the vitreous humor, the gel-like substance inside the eye (see picture below).

 

*Picture: Illustration of the ocular anatomy and intravitreal injection for the treatment of ocular diseases [4]*

One major challenge of developing drug products, which are applied intravitreal, is the requirement for the low viscosity of the drug product solution. Viscosity is the measure of how easily a fluid flows; thicker liquids like honey have high viscosity, while thinner ones like water have low viscosity. It reflects the internal resistance of a liquid's molecules to movement or flow. A high viscosity of the drug product solution in the syringe results in a higher injection force necessary to apply the medication into the eye. The European Pharmacopoeia (EP) provides specific guidelines regarding the viscosity of intravitreal applied biopharmaceuticals to ensure safe and effective injection [5]. 

For this reason the viscosity is a very important measure and is determined several times during the early development stage for every new product. Viscosity is tested under different experiment conditions like temperature and product concentration. In order to reduce development time to the commercial launch of a new drug product and reduce costs for laboratory equipment and personnel, the long term motivation is to predict the viscosity of every new agent without any experiments in the laboratory.

The goal of this work is to examine the impact of different experiment conditions like the temperature and characteristics of the product solution like the product concentration, the protein format, the molecular weight and other factors on the measured viscosity value of the drug product.

According to the literature some of the examined variables are expected to have an impact on the viscosity of the solution. According to the Arrhenius equation, a higher temperature generally decreases viscosity, because molecular movement increases, reducing intermolecular interactions [6]. Also the product concentration might have an impact on the viscosity, because molecules in solution interact more frequently, leading to increased resistance to flow [7] Additionally larger and more complex proteins, such as aggregates or conjugated proteins, tend to increase solution viscosity due to their size and interaction with other molecules in the solution [8], which indicates that also the protein format and the moleculare weight of the product might have an impact on the measured viscosity. Other factors like the replicate number or the instrument should not have any impact on the measurement values, but should be controlled in order to avoid confounding variables, which can potentially reduce the validity of the results. The hypothesis will be tested by systematically varying these factors and measuring viscosity under controlled conditions.



## Data description

The data set I want to explore consists of viscosity data, whereas each observation of the data set corresponds to one measurement value. The data was collected as part of a characterization study for various biopharmaceutical products. These products consist of different types of proteins (IgG2, IgG4, Knob/Hole, DoppelMab), which have different characteristics like molecular weight, isoelectic point or extinction coefficient. In order to determine the effect of product concentration on the viscosity, each product was measured at two different concentrations (10 mg/mL, 62.5 mg/mL).  Furthermore, viscosity was measured at different temperatures (2°C - 40°C) to assess the impact of temperature variations.

The underlying measurements were conducted by laboratory personnel within the formulation development department at Boehringer Ingelheim. The measurement values were documented in the internal Laboratory Information Management System (LIMS) along with additional detailed experimental and contextual information, which also contain all explanatory data like the product characteristics etc. The data is stored in an Oracle SQL database and was extracted using targeted SQL queries.


## Analysis approach

The response variable in our data set is the viscosity measurement value, which corresponds to every observation. The viscosity is documented in the unit mPas. The variable contains continuous numerical values, which are affected by different explanatory data, which might have an impact on it. 

```python
import pandas as pd
df = pd.read_csv("viscosity_data.csv", sep=";")
print(f'Summary statistics for the response variable:  \n  {df['measurement_value_mPas'].describe().T.round(2)}   \n')
```

The response variable "measurement_value_mPas" contains 502 values. This is also the number of the total observations in our data set with each observation corresponding to one measurement value. The mean measured viscosity value is 3.89 mPas, while the standard deviation of 2.99 mPas shows a high distribution of the values within the data set. The highest viscosity value, which was measured within the experiment is 27.87 mPas. The lowest measurement value is 0.54 mPas.

```python
import matplotlib.pyplot as plt
df.boxplot(column=['measurement_value_mPas'])
plt.title('Distribution of measurement values')
plt.show()
```

The box plot above displays the distribution of viscosity measurement values. A notable observation is the presence of numerous data points outside the box, indicating potential outliers as identified by the box plot. These outliers represent values that exceed 1.5 times the interquartile range above the third quartile. Interestingly, all outliers are positioned above the median, indicating a pronounced right skew in the distribution.

Besides the response variable, there are ten additional variables in the data set. Some of the variables are considered as predictors and might have an impact on the response variable, while others should not have any effect on the response variable. In the table below all variables and the first five values are presented:

```python
df = pd.read_csv("viscosity_data.csv", sep=";")
df.head()
```

The following variables are considered as predictors and might have an impact on the response variable

- temperature_c (The temperature at which the measurement was conducted)
- product_concentration_mg_mL (Concentration of the product in the aqueous solution in mg/mL) 
- protein_format (Protein format of the investigated product)
- molecular_weight_kda (Molecular weight of the investigated product in kDa. A measure of the size of the protein)
- extinction_coefficient_l_molcm (Extinction coefficient of the investigated product in L·mol⁻¹·cm⁻¹. A measure of the light absorption ability of the molecule)
- isoelectric_point (Isoelectric point of the investigated product. A measure of the charge of the molecule.)

The following variables are usually not considered as predictors, but might have an impact on the response variable and should be examined in order to reveal confounding effects on the response variable

- entered_on (The date on which the measurement was conducted)
- instrument (Instrument, which was used to measure the viscosity)

The following variables should not be used in the model, but help to understand the data. 

- replicate (Number of replicate. Within each measurement, two individual measurements were conducted as technical replicates)
- product (Internal product name as a unique code)

In order to predict the viscosity we will use linear regression as model type for most of the variables.
