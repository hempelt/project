### Data source

Die verwendeten Daten umfassen Messwerte, die von Mitarbeitenden der Formulierungsentwicklung bei Boehringer Ingelheim im Labor erhoben wurden. Diese Messwerte wurden zusammen mit zusätzlichen, detaillierten Versuchs- und Kontextinformationen im internen Laboratory Information Management System (LIMS) dokumentiert. Die Daten sind in einer Oracle SQL-Datenbank gespeichert und wurden durch gezielte SQL-Abfragen aus der Datenbank extrahiert.

### Data characterisitcs

Die oben erwähnten Messwerte sind Viskositätsdaten, die im Rahmen einer Charakterisierung unterschiedlicher Wirkstoffe erhoben wurden. Bei den Wirkstoffen handelt es sich um unterschiedliche Arten von Proteinen, die in einer wässrigen Lösung mit zusätzlichen stabilisierenden Hilfsstoffen enthalten sind. Die Viskosität beschreibt dabei wie dickflüssig oder zäh die Proteinlösung ist. Jeder Wirkstoff wurde in zwei unterschiedlichen Konzentrationen untersucht, um den Einfluss der Proteinkonzentration auf die Viskosität zu bestimmen. Außerdem wurde die Viskosität bei unterschiedlichen Temperaturen gemessen.

### Research question

Wie groß ist der Einfluss unterschiedlicher Parameter wie Messtemperatur, Proteinkonzentration, Protein Format, Molekulargewicht und anderen Faktoren auf die Viskosität von Arzneimittelprodukten.
