```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import altair as alt
import numpy as np
```

```python
df = pd.read_csv("viscosity_data.csv", sep=";")
df = df.iloc[0:502,[0,4]]
```

```python
df
```

```python
y_label = 'viscosity_mpas'

X = df[['temperature']]
y = df[y_label]
```

```python

chart = (
        alt.Chart(df)
        .mark_circle(size=60)
        .encode(
            x=alt.X('temperature', title="temperature"),
            y=alt.Y('viscosity_mpas', title=y_label),
        )
        .properties(
            width=300,
            height=300
        )
    )

```

```python

chart
```

```python
reg = LinearRegression()
reg.fit(X, y)
y_pred = reg.predict(X)
```

```python
y_pred = reg.predict(X)
```

```python
df['y_pred'] = y_pred
```

```python
df
```

```python
mse = mean_squared_error(df['viscosity_mpas'], y_pred)
print("Mean Squared Error nach Umkehr der Transformation:", mse)
```

```python
df['log_viscosity_mpas'] = np.log(df['viscosity_mpas'])
df[['viscosity_mpas','temperature','y_pred','log_viscosity_mpas']]
```

```python
boxplot = (
        alt.Chart(df)
        .mark_boxplot()
        .encode(
            x=alt.X('temperature', title="temperature"),
            y=alt.Y('log_viscosity_mpas', title='log_viscosity_mpas'),
        )
        .properties(
            width=300,
            height=300
        )
    )

boxplot
```

```python
reglog = LinearRegression()
reglog.fit(X, df['log_viscosity_mpas'])
y_predlog = reglog.predict(X)
```

```python
y_pred_exp=np.exp(y_predlog)
df['y_pre_exp'] = y_pred_exp
df['error']=df['viscosity_mpas'] - df['y_pred']
df.head(40)
```

```python
mse = mean_squared_error(df['log_viscosity_mpas'], y_pred_exp)
print("Mean Squared Error nach Umkehr der Transformation:", mse)
```
